```python
import yfinance as yf
from datetime import datetime
```

```python
start_date = "2025-01-01"
end_date   = "2025-01-31"
```

```python
ticker="CL=F"
```

```python
df_single = yf.download(
    tickers=ticker,
    start=start_date,
    end=end_date,
    interval="1d",
    group_by="ticker",
    auto_adjust=True,
    progress=False
)
```
